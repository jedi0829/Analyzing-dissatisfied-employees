# Identify dissatisfied employees

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def update_vals(v):
    if pd.isnull(v):
        return np.nan
    if v == '-':
        return False
    else:
        return True
# Create a function to convert the text answers into True, False, and NaN. 
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(axis=1, skipna=False)
tafe_resignations['dissatisfied'].value_counts(dropna=False)

# Apply the function created above to TAFE_resignation dataframe to convert the answers, and then count the number of True, False, and NaN. 
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
'dissatisfaction_with_the_department', 
'physical_work_environment', 
'lack_of_recognition', 
'lack_of_job_security', 
'work_location', 
'employment_conditions',
'work_life_balance',
'workload']].any(axis=1, skipna=False)
dete_resignations['dissatisfied'].value_counts(dropna=False)

# count the number of dissatisfied exits at DETE survey using df.any() but no need to use update_vals function because the answers in DETE survey are already in True, False, and NaN.
```

```python
tafe_resignations_up = tafe_resignations.copy()
dete_resignations_up = dete_resignations.copy()
# Make copies of the results to avoid settingwithcopy warning.
```

# Combine the data

```python
# Now the length of employment for each employee is figured out, the number of dissatisfied employees has been identified, now I will proceed and combine the data from two dataframes. 
```

```python
dete_resignations_up['institute'] = 'DETE'
dete_resignations_up['institute'].head()
# add a column named institute to dete_resignations_up with each row containing 'DETE'
```

```python
tafe_resignations_up['institute'] = 'TAFE'
tafe_resignations_up['institute'].head()
# add a column named institute to tafe_resignations_up with each row containing 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], axis=0)
combined.notnull().sum().sort_values()
# Combine the dete_resignations_up, tafe_resignations_up dataframes
# List all non-null values and sort by value
```

```python
combined_updated = combined.dropna(axis=1, thresh=500)
combined_updated.notnull().sum().sort_values()
# Use df.dropna() method to drop any columns with less than 500 non-null values
# Now we only have a handful of columns with enough non-null value for analysis
```

# Clean the Institute_service column

```python
combined_updated['institute_service'].value_counts()
# see what this column contains in data forms
# We need to them into different categories
```

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str').str.extract(r"(\d+)")
combined_updated['institute_service'].value_counts()
# Extract the year value from institute_service column
# If the value is a range, i.e. 11-20, we extract "11" instead of 20, the upper range
# however, this doesn't not impact our analysis because 11-20 will be fallen into the same category in the analysis
```

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('float')
combined_updated['institute_service'].value_counts()
# Convert the column from str to float
```

# Define career stage: 
New: Less than 3 years at a company
Experienced: 3-6 years at a company
Established: 7-10 years at a company
Veteran: 11 or more years at a company

```python
def career_stage(years):
    if years < 3:
        return 'New'
    if 3 <= years < 7:
        return 'Experienced'
    if 7 <= years < 11:
        return 'Established'
    if pd.isnull(years):
        return np.nan
    else:
        return 'Veteran'
# create a function to categorize the respondents' experience
```

```python
combined_updated['service_cat'] = combined_updated['institute_service'].apply(career_stage)
combined_updated['service_cat'].value_counts()
# assign the categorization results to a new column called "service_cat"
```

# Perform some analysis after cleaning

```python
# the goal here is to see if there are any missing values in 'dissatisfied' column

combined_updated['dissatisfied'].value_counts(dropna=False)

# we can see there are 8 NaN values besides True and False
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
combined_updated['dissatisfied'].value_counts()
# here I filled the missing value NaN with False as False accounts for higher portion of the values
```

```python
combined_agg = combined_updated.pivot_table(values='dissatisfied', index='service_cat', aggfunc=np.mean)
combined_agg
# use pivot_table method to aggregate the date to see the dissatisfaction level for each of the service category (new, experienced, established, veteran)
```

```python
combined_agg.plot(kind='bar')
plt.show()
# draw a plot to show each category's percentage of resignation as a result of dissatifaction
```

Now from the chart we know "Established" and "Veteran" groups are experiencing higher percentage in dissatisfied being the reason for quitting.
